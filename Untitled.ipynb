```python
from jax_circuits import *
from topology import *
from scipy.stats import unitary_group
import matplotlib.pyplot as plt
import math
```

```python
num_params = 3
initial_params = random_angles(num_params)
def sample_cost(x):
    return (x**4).sum()

preconditioner_func = hessian_preconditioner(sample_cost, tikhonov_delta=1e-4, learning_rate=0.1)


_, loss_history = gradient_descent_minimize(sample_cost, num_params, preconditioner_func, initial_params=initial_params)
```

```python
plt.plot(loss_history)
plt.yscale('log')
```

```python
num_qubits = 2
depth = theoretical_lower_bound(num_qubits)

anz = Ansatz(num_qubits, 'cz', placements=fill_layers(sequ_layer(num_qubits), depth))
u_target = unitary_group.rvs(2**num_qubits, random_state=1)

anz.circuit().draw(output='mpl')
```

```python
_, loss_history = anz.learn(u_target)
plt.plot(loss_history)
plt.yscale('log')
```

```python
%%time
def cost_func(angles):
    return disc2(anz.unitary(angles), u_target)

initial_angles = random_angles(anz.num_angles, key=random.PRNGKey(10))

preconditioner_func = hessian_preconditioner(cost_func, tikhonov_delta=1e-4, learning_rate=0.01)

angles_history, loss_history = gradient_descent_minimize(cost_func, anz.num_angles, preconditioner_func, initial_params=initial_angles)
```

```python
plt.plot(loss_history)
plt.yscale('log')
```
