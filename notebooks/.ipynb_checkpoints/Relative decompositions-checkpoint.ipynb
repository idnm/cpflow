# Toffoli 4 from relative Toffoli 3

```python
qc_toff4 = QuantumCircuit(5)
qc_toff4.mct([0,1,3], 4)
u_toff4_anc = Operator(qc_toff4.reverse_bits()).data
qc_toff4.draw(output='mpl')
```

```python
d1 = random.uniform(random.PRNGKey(0), minval=0, maxval=2*jnp.pi, shape=(8,))
d2 = random.uniform(random.PRNGKey(1), minval=0, maxval=2*jnp.pi, shape=(4,))

D1 = jnp.diag(jnp.exp(1j*d1))
D2 = jnp.diag(jnp.exp(1j*d2))

U1 = unitary_group.rvs(4, random_state=0)
U2 = unitary_group.rvs(4, random_state=1)

print(jnp.allclose(D1 @ D1.conj().T, jnp.identity(8)))
print(jnp.allclose(D2 @ D2.conj().T, jnp.identity(4)))
```

```python
qc = QuantumCircuit(5)
qc.ccx(0, 1, 2)
qc.unitary(D1, [0,1,2], label='$D_1$')
qc.unitary(U1, [0, 1], label='U_1')
qc.ccx(2, 3, 4)
qc.unitary(D2, [3, 4], label='$D_2$')
qc.unitary(U2, [3, 4], label='$U_2$')
qc.unitary(U1.conj().T, [0, 1], label='$U_1^\dagger$')
qc.unitary(D1.conj().T, [0,1,2], label='$D_1^\dagger$')
qc.ccx(0, 1, 2)
qc.unitary(U2.conj().T, [3, 4], label='$U_2^\dagger$')
qc.unitary(D2.conj().T, [3, 4], label='$D_2^\dagger$')
qc.ccx(2, 3, 4)

u = Operator(qc.reverse_bits()).data
print(jnp.allclose(u, u_toff4_anc))
qc.draw(output='mpl')
```
